<P>THIS MODULE WILL TAKE A EXCEL FILE WITH JUST THE COUNTY AND STATE OF ANY CLIENT LOCATIONS AND BUILD OUT AN EXCEL FILE THAT INCLUDES THE FIPS CODE, POPULATION, LAT/LONG POSITIONS FOR MAP BUILDING, AND INDEXES THE TABLE ON THE FIP COLUMN</P>

```python
#hide
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
#collapse-hide
client=input('Client Name : ')
print('PULLING STATE AND COUNTY LOCATIONS FOR: ' + client.upper())
clientLocations=pd.read_csv('../_data/client_locations/' + client + '.csv')

clientLocations.columns = map(str.lower, clientLocations.columns)
clientLocations
```

```python
#collapse-hide
cenLat=pd.read_csv('../_data/Data_Sources/Census/Geo-Fips.csv')
cenLat
```

```python
#collapse-hide
clm=pd.merge(cenLat, clientLocations, on=['state', 'county'])
clmCol=[ 'fips', 'country', 'state', 'county', 'lat', 'long']
clientDataCleaned=clm[clmCol]
clientDataCleaned
```

```python
#collapse-hide
cen=pd.read_csv('../_data/Data_Sources/Census/Census_2019.csv')
cCol=['state', 'county', 'pop19']
censusData=cen[cCol]
censusData
```

```python
#collapse-hide
ClientCensusCleaned=pd.merge(clientDataCleaned, censusData,  on=['state','county'], how='left')
ClientCensusCleaned.set_index('fips', inplace=True)
clientDataCleaned.shape
```

```python
#collapse-hide
print('THIS IS THE ORIGINAL LOCATIONS FILE FOR:' + client)
clientLocations
```

```python
#collapse-hide
print('THIS IS THE LAT/LONG AND FIPS CODES DATASET')
cenLat.head(3)
```

```python
#collapse-hide
print('THIS IS THE CENSUS POPULATION DATASET')
censusData.head(3)
```

```python
#collapse-hide
print('THIS IS THE NEW CLIENT LOCATIONS FILE')
ClientCensusCleaned
```

```python
#collapse-hide
ClientCensusCleaned.to_csv('../_data/client_Locations/' + client + 'Cleaned.csv')
print('MERGED THE ORIGINAL ' + client + ' LOCATIONS FILE AND ADDED CENSUS DATA, FIPS CODES, AND LAT/LONG')
print ('NEW FILE CAN BE FOUND AT: _DATA/CLIENT_LOCATIONS/' + client + 'Cleaned.csv')
```
